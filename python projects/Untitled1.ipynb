```python
#2
import string

def summarize_letters(input_string):
    # Convert to lowercase
    input_string = input_string.lower()

    # Create a dictionary to hold the frequencies
    frequency_dict = {}

    # Iterate over each character in the string
    for char in input_string:
        # Check if the character is a letter
        if char in string.ascii_lowercase:
            if char in frequency_dict:
                frequency_dict[char] += 1
            else:
                frequency_dict[char] = 1

    return frequency_dict

# Test the function
test_string = "The quick brown fox jumps over the lazy dog!"

# Get the frequency dictionary
frequency_dict = summarize_letters(test_string)

# Print the frequencies
for letter, frequency in frequency_dict.items():
    print(f"{letter}: {frequency}")

# Check if the string contains all the letters of the alphabet
all_letters_present = all(char in frequency_dict for char in string.ascii_lowercase)
if all_letters_present:
    print("The string contains all the letters of the alphabet.")
else:
    print("The string does not contain all the letters of the alphabet.")


```

```python
  #3
  # Given dictionary
  tlds = {'Canada': 'ca', 'United States': 'us', 'Mexico': 'mx'}

  # a) Check if 'Canada' is in the dictionary
  print("Checking if 'Canada' is in the dictionary:")
  if 'Canada' in tlds:
      print("Canada is present.")
  else:
      print("Canada is not present.")

  # b) Check if 'France' is in the dictionary
  print("\nChecking if 'France' is in the dictionary:")
  if 'France' in tlds:
      print("France is present.")
  else:
      print("France is not present.")

  # c) Iterate through key-value pairs and display them
  print("\nIterating through the dictionary:")
  for country, tld in tlds.items():
      print(f"{country}: {tld}")

  # d) Add the key-value pair 'Sweden': 'sw'
  print("\nAdding 'Sweden': 'sw' to the dictionary:")
  tlds['Sweden'] = 'sw'
  print(tlds)

  # e) Update the value for 'Sweden' to 'se'
  print("\nUpdating the value for 'Sweden' to 'se':")
  tlds['Sweden'] = 'se'
  print(tlds)

  # f) Reverse the keys and values using dictionary comprehension
  print("\nReversing keys and values:")
  reversed_tlds = {v: k for k, v in tlds.items()}
  print(reversed_tlds)

  # g) Convert country names to uppercase using dictionary comprehension
  print("\nConverting country names to uppercase:")
  uppercase_countries = {k.upper(): v for k, v in reversed_tlds.items()}
  print(uppercase_countries)
```

```python
#4
import string

def check_palindrome(input_string):
    # Clean the input string: convert to lowercase and remove non-alphanumeric characters
    cleaned_string = ''.join(char for char in input_string.lower() if char.isalnum())

    # Check if the cleaned string is equal to its reverse
    return cleaned_string == cleaned_string[::-1]

# Test cases
print(check_palindrome("radar"))  # True
print(check_palindrome("A man, a plan, a canal, Panama"))  # True
print(check_palindrome("Was it a car or a cat I saw?"))  # True
print(check_palindrome("not a palindrome"))  # False
print(check_palindrome("12321"))  # True
```

```python
#5
import random

# Generate 20 random letters in the range 'a' through 'f'
random_letters = [random.choice('abcdef') for _ in range(20)]
print("Random letters:", random_letters)

#a) Sort the list in ascending order
sorted_asc = sorted(random_letters)
print("Sorted in ascending order:", sorted_asc)

#b) Sort the list in descending order
sorted_desc = sorted(random_letters, reverse=True)
print("Sorted in descending order:", sorted_desc)

#c) Get unique values and sort them in ascending order
unique_sorted = sorted(set(random_letters))
print("Unique values sorted in ascending order:", unique_sorted)
```

```python
#6
# Initialize an empty list to simulate the queue
queue = []

# Enqueue operations
queue.append(3)
queue.append(2)
queue.append(1)

print("Initial queue:", queue)

# Dequeue operations
dequeued_value1 = queue.pop(0)
dequeued_value2 = queue.pop(0)
dequeued_value3 = queue.pop(0)

print("Dequeued values (in FIFO order):", dequeued_value1, dequeued_value2, dequeued_value3)
print("Final queue:", queue)  # Should be empty after dequeuing
```

```python
from collections import deque

# Create an empty deque to represent the queue
queue = deque()

# Enqueue the values 3, 2, and 1
queue.append(3)
print("Queue after enqueue 3:", list(queue))
queue.append(2)
print("Queue after enqueue 2:", list(queue))
queue.append(1)
print("Queue after enqueue 1:", list(queue))

# Dequeue the values and print them to show FIFO order
first_out = queue.popleft()
print("Dequeued:", first_out)
print("Queue after dequeue:", list(queue))

second_out = queue.popleft()
print("Dequeued:", second_out)
print("Queue after dequeue:", list(queue))

third_out = queue.popleft()
print("Dequeued:", third_out)
print("Queue after dequeue:", list(queue))
```
