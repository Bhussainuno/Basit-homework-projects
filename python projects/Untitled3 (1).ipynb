```python
import numpy as np

def initialize_board():
    """
    Initialize the 3x3 Tic-Tac-Toe board with empty strings.

    Returns:
    np.ndarray: A 3x3 NumPy array filled with empty strings ('').
    """
    return np.full((3, 3), '', dtype=str)

def display_board(board):
    """
    Display the current state of the Tic-Tac-Toe board.

    Args:
    board (np.ndarray): The 3x3 NumPy array representing the game board.
    """
    print("\nCurrent Board:")
    print(board)

def check_winner(board, player):
    """
    Check if the specified player has won the game.

    Args:
    board (np.ndarray): The 3x3 NumPy array representing the game board.
    player (str): The player's mark ('X' or 'O') to check for winning condition.

    Returns:
    bool: True if the player has won, False otherwise.
    """
    # Check rows and columns for three consecutive marks
    for i in range(3):
        if all(board[i, :] == player) or all(board[:, i] == player):
            return True

    # Check diagonals
    if all(np.diag(board) == player) or all(np.diag(np.fliplr(board)) == player):
        return True

    return False

def check_draw(board):
    """
    Check if the game is a draw (all squares filled).

    Args:
    board (np.ndarray): The 3x3 NumPy array representing the game board.

    Returns:
    bool: True if the game is a draw, False otherwise.
    """
    return '' not in board

def check_game_state(board):
    """
    Check the current state of the Tic-Tac-Toe game.

    Args:
    board (np.ndarray): The 3x3 NumPy array representing the game board.

    Returns:
    str: Message indicating current game state:
         - "Player 'X' is winning!" if 'X' has won.
         - "Player 'O' is winning!" if 'O' has won.
         - "It's a draw!" if the game is a draw.
         - "Game in progress..." if no winner and no draw yet.
    """
    if check_winner(board, 'X'):
        return "Player 'X' is winning!"
    elif check_winner(board, 'O'):
        return "Player 'O' is winning!"
    elif check_draw(board):
        return "It's a draw!"
    else:
        return "Game in progress..."

def get_player_move(board, player):
    """
    Get the player's move (row and column indices).

    Args:
    board (np.ndarray): The 3x3 NumPy array representing the game board.
    player (str): The current player's mark ('X' or 'O').

    Returns:
    tuple: The row and column indices chosen by the player.
    """
    while True:
        try:
            row = int(input(f"\nPlayer '{player}', enter the row index (0, 1, or 2): "))
            col = int(input(f"Player '{player}', enter the column index (0, 1, or 2): "))

            if 0 <= row <= 2 and 0 <= col <= 2:
                if board[row, col] == '':
                    return row, col
                else:
                    print("That square is already taken. Try again.")
            else:
                print("Row and column indices must be between 0 and 2. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def main():
    """
    Main function to play the Tic-Tac-Toe game between two human players.
    """
    board = initialize_board()  # Initialize the game board
    current_player = 'X'  # Player 1 starts with 'X'

    while True:
        display_board(board)  # Display current board state

        # Get player's move
        row, col = get_player_move(board, current_player)
        board[row, col] = current_player

        # Check the current game state
        game_state = check_game_state(board)
        if game_state != "Game in progress...":
            display_board(board)
            print(f"\n{game_state}")
            break

        # Switch to the other player
        current_player = 'O' if current_player == 'X' else 'X'

if __name__ == "__main__":
    main()
```

```python
#7
import numpy as np
import random

def initialize_board():
    """
    Initialize the 3x3 Tic-Tac-Toe board with empty strings.

    Returns:
    np.ndarray: A 3x3 NumPy array filled with empty strings ('').
    """
    return np.full((3, 3), '', dtype=str)

def display_board(board):
    """
    Display the current state of the Tic-Tac-Toe board.

    Args:
    board (np.ndarray): The 3x3 NumPy array representing the game board.
    """
    print("\nCurrent Board:")
    print(board)

def check_winner(board, player):
    """
    Check if the specified player has won the game.

    Args:
    board (np.ndarray): The 3x3 NumPy array representing the game board.
    player (str): The player's mark ('X' or 'O') to check for winning condition.

    Returns:
    bool: True if the player has won, False otherwise.
    """
    # Check rows and columns for three consecutive marks
    for i in range(3):
        if all(board[i, :] == player) or all(board[:, i] == player):
            return True

    # Check diagonals
    if all(np.diag(board) == player) or all(np.diag(np.fliplr(board)) == player):
        return True

    return False

def check_draw(board):
    """
    Check if the game is a draw (all squares filled).

    Args:
    board (np.ndarray): The 3x3 NumPy array representing the game board.

    Returns:
    bool: True if the game is a draw, False otherwise.
    """
    return '' not in board

def check_game_state(board):
    """
    Check the current state of the Tic-Tac-Toe game.

    Args:
    board (np.ndarray): The 3x3 NumPy array representing the game board.

    Returns:
    str: Message indicating current game state:
         - "Player 'X' is winning!" if 'X' has won.
         - "Player 'O' is winning!" if 'O' has won.
         - "It's a draw!" if the game is a draw.
         - "Game in progress..." if no winner and no draw yet.
    """
    if check_winner(board, 'X'):
        return "Player 'X' is winning!"
    elif check_winner(board, 'O'):
        return "Player 'O' is winning!"
    elif check_draw(board):
        return "It's a draw!"
    else:
        return "Game in progress..."

def get_player_move(board, player):
    """
    Get the player's move (row and column indices).

    Args:
    board (np.ndarray): The 3x3 NumPy array representing the game board.
    player (str): The current player's mark ('X' or 'O').

    Returns:
    tuple: The row and column indices chosen by the player.
    """
    while True:
        try:
            row = int(input(f"\nPlayer '{player}', enter the row index (0, 1, or 2): "))
            col = int(input(f"Player '{player}', enter the column index (0, 1, or 2): "))

            if 0 <= row <= 2 and 0 <= col <= 2:
                if board[row, col] == '':
                    return row, col
                else:
                    print("That square is already taken. Try again.")
            else:
                print("Row and column indices must be between 0 and 2. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def get_computer_move(board, computer):
    """
    Get the computer's move (random selection of an empty cell).

    Args:
    board (np.ndarray): The 3x3 NumPy array representing the game board.
    computer (str): The computer's mark ('X' or 'O').

    Returns:
    tuple: The row and column indices chosen by the computer.
    """
    empty_cells = np.where(board == '')
    idx = random.choice(range(len(empty_cells[0])))
    row = empty_cells[0][idx]
    col = empty_cells[1][idx]
    return row, col

def main():
    """
    Main function to play the Tic-Tac-Toe game between a human player and the computer.
    """
    board = initialize_board()  # Initialize the game board

    # Determine player and computer marks
    player_choice = input("Do you want to be 'X' or 'O'? ").upper()
    computer_choice = 'X' if player_choice == 'O' else 'O'

    current_player = 'X'  # Player 1 starts with 'X'

    # Decide who starts first
    if player_choice == 'O':
        print("Player 'O' goes first.")
    else:
        print("You ('X') go first.")

    while True:
        display_board(board)  # Display current board state

        # Human player's move
        if current_player == player_choice:
            row, col = get_player_move(board, current_player)
        # Computer's move
        else:
            row, col = get_computer_move(board, current_player)
            print(f"\nComputer ({current_player}) chooses row {row} and column {col}.")

        # Make the move on the board
        board[row, col] = current_player

        # Check the current game state
        game_state = check_game_state(board)
        if game_state != "Game in progress...":
            display_board(board)
            print(f"\n{game_state}")
            break

        # Switch to the other player
        current_player = 'O' if current_player == 'X' else 'X'

if __name__ == "__main__":
    main()
```
